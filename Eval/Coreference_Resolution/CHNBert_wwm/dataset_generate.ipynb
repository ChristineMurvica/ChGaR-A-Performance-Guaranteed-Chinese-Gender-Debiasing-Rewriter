## 下面是数据转换代码-Pro

```python
f1 = open('./extradata/my_dataset/re_wri_pro_stereotyped_type1_dev_ch.txt',"r",encoding='utf-8') #re_wri_
lines = f1.readlines()
p_dev = pd.DataFrame(columns=("idx","sen"))
a = 0
for line in lines:
    csvRow = line.split(" ")
    s = {}
    s["idx"]= int(csvRow[0])
    s["sen"]= str(csvRow[1])
    p_dev.loc[a] = s
    a += 1
p_dev=p_dev.drop("idx", axis=1)

f2 = open('./extradata/my_dataset/re_wri_pro_stereotyped_type1_test_ch.txt',"r",encoding='utf-8') #re_wri_
lines = f2.readlines()
p_test = pd.DataFrame(columns=("idx","sen"))
a = 0
for line in lines:
    csvRow = line.split(" ")
    s = {}
    s["idx"]= int(csvRow[0])
    s["sen"]= str(csvRow[1])
    p_test.loc[a] = s
    a += 1
p_test=p_test.drop("idx", axis=1)

p=pd.concat([p_dev,p_test],ignore_index=True)
p
```

## all_dict

```python
all_dict={}

#pro
c=0
for i in range(p.shape[0]):
    temp=re.findall(r"[[](.*?)[]]", p.iloc[i,0])
    if len(temp)!=2:
        continue
    for j in range(len(temp)):
        if not temp[j] in all_dict:
            all_dict[str(temp[j])]=c
            c+=1
            
#anti
for i in range(a.shape[0]):
    temp=re.findall(r"[[](.*?)[]]", a.iloc[i,0])
    if len(temp)!=2:
        continue
    for j in range(len(temp)):
        if not temp[j] in all_dict:
            all_dict[str(temp[j])]=c
            c+=1
all_dict
```

```python
all_return=[]
print(p.shape[0])
for i in range(p.shape[0]):
    result={}
    t_dict={}
    sen=p.iloc[i,0]
    temp=re.findall(r"[[](.*?)[]]", sen)
    if len(temp)!=2:
        continue
    #print(i,temp[1],len(temp))
    if temp[0]=='他'or temp[0]=='她':        
        t_dict["span2_index"]=all_dict[str(temp[0])]
        t_dict["span1_index"]=all_dict[str(temp[1])]
        t_dict["span1_text"]=temp[1]
        t_dict["span2_text"]=temp[0]
        result["target"]=t_dict
        result["idx"]=i
        result["label"]="true"
        result["text"]=str(sen)
        all_return.append(result)
        continue
    elif temp[1]=='他'or temp[1]=='她':
        #print(1)
        t_dict["span2_index"]=all_dict[str(temp[1])]
        t_dict["span1_index"]=all_dict[str(temp[0])]
        t_dict["span1_text"]=temp[0]
        t_dict["span2_text"]=temp[1]
        result["target"]=t_dict
        result["idx"]=i
        result["label"]="true"
        result["text"]=str(sen)
        all_return.append(result)
```

```python
print(a.shape[0])
for i in range(a.shape[0]):
    result={}
    t_dict={}
    sen=a.iloc[i,0]
    temp=re.findall(r"[[](.*?)[]]", sen)
    if len(temp)!=2:
        continue
    #print(i,temp[1],len(temp))
    if temp[0]=='他'or temp[0]=='她':        
        t_dict["span2_index"]=all_dict[str(temp[0])]
        t_dict["span1_index"]=all_dict[str(temp[1])]
        t_dict["span1_text"]=temp[1]
        t_dict["span2_text"]=temp[0]
        result["target"]=t_dict
        result["idx"]=i
        result["label"]="false"
        result["text"]=str(sen)
        all_return.append(result)
        continue
    elif temp[1]=='他'or temp[1]=='她':
        #print(1)
        t_dict["span2_index"]=all_dict[str(temp[1])]
        t_dict["span1_index"]=all_dict[str(temp[0])]
        t_dict["span1_text"]=temp[0]
        t_dict["span2_text"]=temp[1]
        result["target"]=t_dict
        result["idx"]=i
        result["label"]="false"
        result["text"]=str(sen)
        all_return.append(result)
```
